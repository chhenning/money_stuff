```python
import os
import re
import sqlite3
import time

import spacy
from spacy import displacy

import torch

from money_stuff import get_all_ml_articles

```

```python
def extract_ents(doc, doc_id):
    return [
        {
            "doc_id": doc_id,
            "text": ent.text,
            "label": ent.label_,
            "start": ent.start_char,
            "end": ent.end_char,
        }
        for ent in doc.ents
    ]
```

```python
articles = get_all_ml_articles()
len(articles)

```

```python
print("MPS available:", torch.backends.mps.is_available())
print("MPS built:", torch.backends.mps.is_built())

spacy.require_gpu()  # must come before spacy.load()
nlp = spacy.load("en_core_web_trf")

num = 100
ids = [a['id'] for a in articles]
ml_texts = [a['ml_text'] for a in articles]
texts = ml_texts[:num]

start = time.time()

docs = list(nlp.pipe(texts, batch_size=4))
total = 0
for id, doc in enumerate(docs):
    total += len(extract_ents(doc, ids[id]))

print(f"Num Entities: {total}")
print(f"Time: {time.time() - start}")

# doc = nlp(articles[0]['ml_text'])
# entities = extract_ents(doc, articles[0]['id'])
```
